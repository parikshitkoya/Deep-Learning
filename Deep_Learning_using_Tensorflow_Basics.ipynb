<center><h1><strong>Basics of Deep Learning using Tensorflow</strong></h1></center>

<strong>Importing Libraries</strong>

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
```

<strong>Creating a single layer Neural Network</strong>

```python
model = tf.keras.Sequential([keras.layers.Dense(units= 1, input_shape = [1])])
```

<strong>Compiling the Neural Network for training</strong>

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

<strong>Generating the data for training</strong>

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

<strong>Fitting the Neural Network</strong>

```python
model.fit(xs, ys, epochs=500)
```

<strong>Making some predictions</strong>

```python
print(model.predict([6.0]))
```

```python
print(model.predict([9.0]))
```
